```python
import pandas as pd

uri = "https://gist.githubusercontent.com/guilhermesilveira/2d2efa37d66b6c84a722ea627a897ced/raw/10968b997d885cbded1c92938c7a9912ba41c615/tracking.csv"
dados = pd.read_csv(uri)
dados.head
```

```python
mapa = {
    "home" : "principal",
    "how_it_works" : "como_funciona",
    "contact" : "contato",
    "bought" : "comprou"
}
dados = dados.rename(columns = mapa)
```

```python
x = dados[["principal","como_funciona","contato"]]
x.head()
```

```python
y = dados["comprou"]
y.head()
```

```python
dados.shape
```

```python
treino_x = x[:75]
treino_y = y[:75]
teste_x = x[75:]
teste_y = y[75:]

print("treinamento foi feito até 75; teste feito de 75 em diante (material tem 99 itens)")
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

modelo = LinearSVC()
modelo.fit(treino_x, treino_y)
previsao = modelo.predict(teste_x)

taxa_de_acerto = accuracy_score(teste_y, previsao)
print("a acurácia é de %.2f%%" %taxa_de_acerto)
```
